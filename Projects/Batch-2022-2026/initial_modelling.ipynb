```python
# stopwords list that needs to be excluded from the data
stopwordlist = ['a', 'about', 'above', 'after', 'again', 'ain', 'all', 'am', 'an',
             'and','any','are', 'as', 'at', 'be', 'because', 'been', 'before',
             'being', 'below', 'between','both', 'by', 'can', 'd', 'did', 'do',
             'does', 'doing', 'down', 'during', 'each','few', 'for', 'from',
             'further', 'had', 'has', 'have', 'having', 'he', 'her', 'here',
             'hers', 'herself', 'him', 'himself', 'his', 'how', 'i', 'if', 'in',
             'into','is', 'it', 'its', 'itself', 'just', 'll', 'm', 'ma',
             'me', 'more', 'most','my', 'myself', 'now', 'o', 'of', 'on', 'once',
             'only', 'or', 'other', 'our', 'ours','ourselves', 'out', 'own', 're',
             's', 'same', 'she', "shes", 'should', "shouldve",'so', 'some', 'such',
             't', 'than', 'that', "thatll", 'the', 'their', 'theirs', 'them',
             'themselves', 'then', 'there', 'these', 'they', 'this', 'those',
             'through', 'to', 'too','under', 'until', 'up', 've', 'very', 'was',
             'we', 'were', 'what', 'when', 'where','which','while', 'who', 'whom',
             'why', 'will', 'with', 'won', 'y', 'you', "youd","youll", "youre",
             "youve", 'your', 'yours', 'yourself', 'yourselves']
```

```python
# removing stopwoords from the tweet text
STOPWORDS = set(stopwordlist)

def clean_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

cyberbulling['tweet_text'] = cyberbulling['tweet_text'].apply(lambda text: clean_stopwords(text))
```

```python
# cleaning and removing punctuations

import string
english_puctuations = string.punctuation

def clean_puctuations(text):
    translator = str.maketrans('','', english_puctuations)
    return text.translate(translator)

cyberbulling['tweet_text'] = cyberbulling['tweet_text'].apply(lambda x : clean_puctuations(x))
```

```python
cyberbulling['tweet_text'][20]
```

```python
# cleaning and removing repeating characters

def clean_repeating_characters(text):
    return re.sub(r'(.)1+', r'1', text)

cyberbulling['tweet_text'] = cyberbulling['tweet_text'].apply(lambda x: clean_repeating_characters(x))
```

```python
cyberbulling['tweet_text'][55]
```

```python
# cleaning and removing URLs

def clean_URLs(text):
    return re.sub(r"((www.[^s]+)|(http\S+))","",text)

cyberbulling['tweet_text'] = cyberbulling['tweet_text'].apply(lambda x : clean_URLs(x))
```

```python
cyberbulling['tweet_text'][280]
```

```python
# cleaning and removing numeric data

def clean_numeric(text):
    return re.sub('[0-9]+', '', text)
    
cyberbulling['tweet_text'] = cyberbulling['tweet_text'].apply(lambda x: clean_numeric(x))
```

```python
cyberbulling['tweet_text'][223]
```

```python
# Tokenization of tweet text

from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer('\w+')
cyberbulling['tweet_text'] = cyberbulling['tweet_text'].apply(tokenizer.tokenize)
```

```python
cyberbulling['tweet_text'][24]
```

```python
# stemming

import nltk

st = nltk.PorterStemmer()

def text_stemming(text):
    text = [st.stem(word) for word in text]
    return text

cyberbulling['tweet_text'] = cyberbulling['tweet_text'].apply(lambda x: text_stemming(x))
```

```python
cyberbulling['tweet_text'][35]
```

```python
# lemmatization
nltk.download('wordnet')
lm = nltk.WordNetLemmatizer()

def text_lemmatization(text):
    text = [lm.lemmatize(word) for word in text]
    return text

cyberbulling['tweet_text'] = cyberbulling['tweet_text'].apply(lambda x: text_lemmatization(x))
```

```python
# Transforming the data using TF-IDF Vectorizer
vectoriser = TfidfVectorizer(ngram_range=(1,2), max_features= 500000)
vectoriser.fit(X_train)
print("No. of feature words: ",len(vectoriser.get_feature_names_out()))
```

```python
# transforming the data
X_train = vectoriser.transform(X_train)
X_test = vectoriser.transform(X_test)
```

```python
# Model
from sklearn.svm import SVC
svm_model_linear = SVC(kernel= 'linear', C = 1).fit(X_train, y_train)
svm_predictions  = svm_model_linear.predict(X_test)
accuracy = svm_model_linear.score(X_test, y_test)
print(accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generate sample data
X, y = make_classification(n_samples=3000, n_features=70, random_state=42)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Initialize Random Forest classifier
random_forest_classifier = RandomForestClassifier(n_estimators=45, random_state=42)

# Train the model
random_forest_classifier.fit(X_train, y_train)

# Make predictions
y_pred = random_forest_classifier.predict(X_test)

# Calculate accuracy
accuracy1 = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy1)
```

```python
from sklearn.metrics import mean_squared_error
mse1 = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse1)
```
